```python
import pandas as pd
import snscrape.modules.twitter as sntwitter

# Creating list to append tweet data to
attributes_container = []

# Defining date range for the past 10 years
start_date = pd.Timestamp('2021-04-01')
end_date = pd.Timestamp('2021-04-01')

# Using TwitterSearchScraper to scrape data and append tweets to list
for i, day in enumerate(pd.date_range(start=start_date, end=end_date, freq='D')):
    if i > 3650: # only scrape data for the past 10 years
        break
    query = 'msft since:{} until:{} lang:en exclude:retweets -filter:replies'.format(day.strftime('%Y-%m-%d'), (day+pd.Timedelta(days=1)).strftime('%Y-%m-%d'))
    tweets = sntwitter.TwitterSearchScraper(query).get_items()
    j = 0
    for tweet in tweets:
        if j >= 10: # limit to 5000 tweets per day
            break
        attributes_container.append([tweet.user.username, tweet.date, tweet.likeCount, tweet.sourceLabel, tweet.rawContent])
        j += 1
    
# Creating a dataframe to load the list
tweets_df = pd.DataFrame(attributes_container, columns=["User", "Date Created", "Number of Likes", "Source of Tweet", "Tweet"])
```
